## 1. Setup and Imports

Import necessary libraries and load the reconstruction code.


```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import sys

# Add src directory to path
sys.path.insert(0, str(Path.cwd() / 'src'))

# Import main reconstruction functions
from main import *

print('✅ Setup complete!')
```

## 2. Load Data

Load input data, ground truth, and reconstruction results.


```python
# Load data files
input_data = np.load('data/gt_output.npy')
gt_output = np.load('data/gt_output.npy')
recon_output = np.load('data/recon_output.npy')

print(f'Input shape: {input_data.shape if "input_data" in locals() else "N/A"}')
print(f'GT output shape: {gt_output.shape if "gt_output" in locals() else "N/A"}')
print(f'Recon output shape: {recon_output.shape if "recon_output" in locals() else "N/A"}')
```

## 3. Visualize Results

Compare input, ground truth, and reconstruction.


```python
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Handle 2D or 3D data
def get_slice(data):
    if data.ndim == 3:
        return data[data.shape[0]//2]  # Middle slice
    return data

if 'input_data' in locals():
    axes[0].imshow(get_slice(input_data), cmap='gray')
    axes[0].set_title('Input', fontsize=14, fontweight='bold')
    axes[0].axis('off')

if 'gt_output' in locals():
    axes[1].imshow(get_slice(gt_output), cmap='gray')
    axes[1].set_title('Ground Truth', fontsize=14, fontweight='bold')
    axes[1].axis('off')

if 'recon_output' in locals():
    axes[2].imshow(get_slice(recon_output), cmap='gray')
    axes[2].set_title('Reconstruction', fontsize=14, fontweight='bold')
    axes[2].axis('off')

plt.tight_layout()
plt.show()
```

## 4. Quantitative Metrics

Evaluate reconstruction quality.


```python
from skimage.metrics import peak_signal_noise_ratio, structural_similarity

if 'gt_output' in locals() and 'recon_output' in locals():
    # Ensure same shape
    if gt_output.shape == recon_output.shape:
        psnr = peak_signal_noise_ratio(gt_output, recon_output, 
                                       data_range=gt_output.max() - gt_output.min())
        ssim_val = structural_similarity(gt_output, recon_output,
                                        data_range=gt_output.max() - gt_output.min())
        
        print(f'📊 PSNR: {psnr:.4f} dB')
        print(f'📊 SSIM: {ssim_val:.4f}')
        print(f'📊 Expected PSNR: 62.77 dB')
        print(f'📊 Expected SSIM: N/A (1D light curve)')
    else:
        print('⚠️ Shape mismatch between GT and reconstruction')
else:
    print('⚠️ GT or reconstruction data not available')
```
